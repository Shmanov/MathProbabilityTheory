```python
import numpy as np
from math import factorial
```

#### 1. –î–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –∏–∑ –≤—ã–±–æ—Ä–∫–∏ –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. –ü–æ—Å—á–∏—Ç–∞—Ç—å (–∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –Ω–∞–ø–æ–¥–æ–±–∏–µ std, var, mean) —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ, —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ, —Å–º–µ—â–µ–Ω–Ω—É—é –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.

```python
X = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
```

```python
# —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ
mean = X.sum() / X.size
mean
```

```python
# —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 
Y = np.array(list(map(lambda x:(x - m)**2, X)))
var = Y.sum() / Y.size
std = np.sqrt(var)
std 
```

```python
# —Å–º–µ—â–µ–Ω–Ω–∞—è –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–π
var_n = Y.sum() / (Y.size - 1)
var, var_n
```

#### 2. –í –ø–µ—Ä–≤–æ–º —è—â–∏–∫–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è 8 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 - –±–µ–ª—ã–µ. –í–æ –≤—Ç–æ—Ä–æ–º —è—â–∏–∫–µ - 12 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 –±–µ–ª—ã—Ö. –ò–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ –≤—ã—Ç–∞—Å–∫–∏–≤–∞—é—Ç —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –¥–≤–∞ –º—è—á–∞, –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ - 4. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ 3 –º—è—á–∞ –±–µ–ª—ã–µ?

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
p1 = combinations(5, 2) / combinations(8, 2) * combinations(5, 1) * combinations(7, 3) / combinations(12, 4)
p2 = combinations(5, 1) * combinations(3, 1) / combinations(8, 2) * combinations(5, 2) * combinations(7, 2) / combinations(12, 4)
p3 = combinations(3, 2) / combinations(8, 2) * combinations(5, 3) * combinations(7, 1) / combinations(12, 4)
p1 + p2 + p3
```

#### 3. –ù–∞ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ –±–∏–∞—Ç–ª–æ–Ω—É –æ–¥–∏–Ω –∏–∑ —Ç—Ä–µ—Ö —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ —Å—Ç—Ä–µ–ª—è–µ—Ç –∏ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –º–∏—à–µ–Ω—å. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞ —Ä–∞–≤–Ω–∞ 0.9, –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ ‚Äî 0.8, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–≥–æ ‚Äî 0.6. –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω: a). –ø–µ—Ä–≤—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º –±). –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º –≤). —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º.

$$P(B\:|\:A) = \frac{P(B) \cdot P(A\:|\:B)}{P(A)}$$

```python
P_AB_1 = 0.9
P_AB_2 = 0.8 
P_AB_3 = 0.6 
P_B = 1/3
P_A = P_B * (P_AB_1 + P_AB_2 + P_AB_3)

P_BA_1 = P_B * P_AB_1 / P_A 
P_BA_2 = P_B * P_AB_2 / P_A
P_BA_3 = P_B * P_AB_3 / P_A
P_BA_1, P_BA_2, P_BA_3
```

#### 4. –í —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç—ã A –∏ B –ø–æ—Å—Ç—É–ø–∏–ª–æ —Ä–∞–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∞ –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç C —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ—Å—Ç—É–ø–∏–ª–æ —Å—Ç–æ–ª—å–∫–æ –∂–µ, —Å–∫–æ–ª—å–∫–æ –Ω–∞ A –∏ B –≤–º–µ—Å—Ç–µ. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ A —Å–¥–∞—Å—Ç –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é, —Ä–∞–≤–Ω–∞ 0.8. –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ B —ç—Ç–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 0.7, –∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ C - 0.9. –°—Ç—É–¥–µ–Ω—Ç —Å–¥–∞–ª –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–Ω —É—á–∏—Ç—Å—è: a). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A –±). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B –≤). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C?

```python
P_AB_A = 0.8
P_AB_B = 0.7 
P_AB_C = 0.9

# –ø—É—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Å —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ –±—É–¥–µ—Ç x, —Ç–æ–≥–¥–∞ x + x + 2x = 1. x = 1|4

# –∞–ø—Ä–∏–æ—Ä–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
P_B_A = P_B_B= 1/4
P_B_C = 2 * 1/4

# –ø–æ–ª–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
P_A = P_B_A * P_AB_A + P_B_B * P_AB_B + P_B_C * P_AB_C

# ùëÉ(ùêµ|ùê¥)=ùëÉ(ùêµ)‚ãÖùëÉ(ùê¥|ùêµ)/ùëÉ(ùê¥)

P_BA_A = P_B_A * P_AB_A / P_A
P_BA_B = P_B_B * P_AB_B / P_A
P_BA_C = P_B_C * P_AB_C / P_A

P_BA_A, P_BA_B, P_BA_C
```

#### 5. –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Ç—Ä–µ—Ö –¥–µ—Ç–∞–ª–µ–π. –î–ª—è –ø–µ—Ä–≤–æ–π –¥–µ—Ç–∞–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–π—Ç–∏ –∏–∑ —Å—Ç—Ä–æ—è –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü —Ä–∞–≤–Ω–∞ 0.1, –¥–ª—è –≤—Ç–æ—Ä–æ–π - 0.2, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–π - 0.25. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –≤—ã–π–¥—É—Ç –∏–∑ —Å—Ç—Ä–æ—è: –∞). –≤—Å–µ –¥–µ—Ç–∞–ª–∏ –±). —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏ –≤). —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å –≥). –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π?

```python
p_1 = 0.1 * 0.2 * 0.25

p_2 = 0.9 * 0.2 * 0.25
p_3 = 0.8 * 0.1 * 0.25
p_4 = 0.75 * 0.1 * 0.2

p_5 = 0.1 * 0.8 * 0.75
p_6 = 0.2 * 0.9 * 0.75
p_7 = 0.25 * 0.9 * 0.8

p_1, p_2 + p_3 + p_4, p_5 + p_6 + p_7, (p_2 + p_3 + p_4) + (p_5 + p_6 + p_7)
```
